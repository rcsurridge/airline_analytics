```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.ml import Pipeline
from pyspark.ml.feature import RegexTokenizer, StopWordsRemover, Normalizer, OneHotEncoder, StringIndexer, VectorAssembler
from pyspark.ml.feature import CountVectorizer, HashingTF, IDF, Word2Vec
from pyspark.ml.classification import LogisticRegression, DecisionTreeClassifier, NaiveBayes, RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
import sparknlp
from sparknlp.pretrained import PretrainedPipeline
from sparknlp.base import DocumentAssembler, Finisher
from sparknlp.annotator import Tokenizer, Normalizer, LemmatizerModel, StopWordsCleaner, NGramGenerator, PerceptronModel
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi'] = 150
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
import numpy as np
import pandas as pd
```

```python
df = spark.read.csv("gs://msca-bdp-student-gcs/Group7_Final_Project/airline_reviews/airline_reviews_preprocessed.csv", 
                                     header = True, inferSchema = True)
```

```python
df.count()
```

```python
df.printSchema()
```

```python
df2 = df.where('overall_score is not null')
df3 = df2.where('review is not null')
```

```python
df3.count()
```

```python
df3 = df3.withColumn('sentiment', lit(None))
df3 = df3.withColumn('sentiment', when((col('overall_score') >= 1) & (col('overall_score') <= 5), "Negative").otherwise(col('sentiment')))
df3 = df3.withColumn('sentiment', when((col('overall_score') >= 6) & (col('overall_score') <= 10), "Positive").otherwise(col('sentiment')))
sentiment_dist = df3.groupBy('sentiment').count()
sentiment_dist = sentiment_dist.withColumn('pct', col('count')/116145)
sentiment_dist.toPandas()
```

```python
df3[['review', 'sentiment']].toPandas()
```

## Exploratory Data Analysis

```python
sentiment_by_year_flown = df3.groupBy(['year_flown', 'sentiment']).count().orderBy(['year_flown', 'sentiment']).toPandas()
sentiment_by_year_flown_2 = pd.merge(sentiment_by_year_flown, year_flown_dist.rename(columns = {'count' : 'total'})[['year_flown', 'total']], on = 'year_flown', how = 'inner')
sentiment_by_year_flown_2['pct'] = sentiment_by_year_flown_2['count']/sentiment_by_year_flown_2['total']
sentiment_by_year_flown_2_neg = sentiment_by_year_flown_2[sentiment_by_year_flown_2['sentiment'] == "Negative"]
sentiment_by_year_flown_2_neg['rel_pct'] = sentiment_by_year_flown_2_neg['count']/sentiment_by_year_flown_2_neg['count'].sum()
sentiment_by_year_flown_2_pos = sentiment_by_year_flown_2[sentiment_by_year_flown_2['sentiment'] == "Positive"]
sentiment_by_year_flown_2_pos['rel_pct'] = sentiment_by_year_flown_2_pos['count']/sentiment_by_year_flown_2_pos['count'].sum()
```

### Average ratings by time

```python
avg_ratings_by_year_flown = df3.groupBy(['year_flown']).agg(mean(col('entertainment_rating')).alias('entertainment_rating'),
                                                            mean(col('food_rating')).alias('food_rating'),
                                                            mean(col('ground_service_rating')).alias('ground_service_rating'),
                                                            mean(col('seat_comfort_rating')).alias('seat_comfort_rating'),
                                                            mean(col('service_rating')).alias('service_rating'),
                                                            mean(col('value_rating')).alias('value_rating'),
                                                            mean(col('wifi_rating')).alias('wifi_rating')).orderBy(['year_flown'])
```

```python
cols = avg_ratings_by_year_flown.columns[1:]
avg_ratings_by_year_flown_2 = avg_ratings_by_year_flown.selectExpr('year_flown', "stack({}, {})".format(len(cols), ', '.join(("'{}', {}".format(i, i) for i in cols))))
avg_ratings_by_year_flown_2 = avg_ratings_by_year_flown_2.withColumnRenamed('col0', 'rating').withColumnRenamed('col1', 'average_score')
avg_ratings_by_year_flown_3 = avg_ratings_by_year_flown_2.toPandas()
```

```python
avg_ratings_by_year_flown_3['rating'] = np.where(avg_ratings_by_year_flown_3['rating'] == 'entertainment_rating', 'Entertainment', avg_ratings_by_year_flown_3['rating'])
avg_ratings_by_year_flown_3['rating'] = np.where(avg_ratings_by_year_flown_3['rating'] == 'food_rating', 'Food', avg_ratings_by_year_flown_3['rating'])
avg_ratings_by_year_flown_3['rating'] = np.where(avg_ratings_by_year_flown_3['rating'] == 'ground_service_rating', 'Ground Service', avg_ratings_by_year_flown_3['rating'])
avg_ratings_by_year_flown_3['rating'] = np.where(avg_ratings_by_year_flown_3['rating'] == 'seat_comfort_rating', 'Seat Comfort', avg_ratings_by_year_flown_3['rating'])
avg_ratings_by_year_flown_3['rating'] = np.where(avg_ratings_by_year_flown_3['rating'] == 'service_rating', 'Service', avg_ratings_by_year_flown_3['rating'])
avg_ratings_by_year_flown_3['rating'] = np.where(avg_ratings_by_year_flown_3['rating'] == 'value_rating', 'Value', avg_ratings_by_year_flown_3['rating'])
avg_ratings_by_year_flown_3['rating'] = np.where(avg_ratings_by_year_flown_3['rating'] == 'wifi_rating', 'Wifi', avg_ratings_by_year_flown_3['rating'])
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data = avg_ratings_by_year_flown_3.rename(columns = {'rating': 'Rating'}), x = 'year_flown', y = 'average_score', hue = 'Rating')
plt.xlabel('Trip Year')
plt.ylabel('Average Rating')
plt.show()
```
