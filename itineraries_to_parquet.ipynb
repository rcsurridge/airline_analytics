```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("CSVtoParquet").getOrCreate()

# Read CSV file into a DataFrame
csv_file_path = "gs://msca-bdp-student-gcs/Group7_Final_Project/flight_pricing/itineraries.csv"
df = spark.read.csv(csv_file_path, header=True, inferSchema=True)

# Save DataFrame as Parquet
parquet_output_path = "gs://msca-bdp-student-gcs/Group7_Final_Project/flight_pricing/itineraries.parquet"
df.write.parquet(parquet_output_path)

# Stop Spark session
spark.stop()
```
