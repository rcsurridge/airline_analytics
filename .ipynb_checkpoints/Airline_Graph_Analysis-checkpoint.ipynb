```python
from graphframes import *
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df_flight_status = spark.read.parquet('gs://msca-bdp-student-gcs/Group7_Final_Project/flight_status/flight_status.parquet',
                                      header = True, inferSchema = True)
```

```python
df_flight_status.printSchema()
```

```python
df_flight_status2 = df_flight_status[['FlightDate',
'Airline',
'Origin',
'Dest',
'Cancelled',
'Diverted',
'Distance',
'DepDelayMinutes',
'Year',
'Quarter',
'Month',
'DayofMonth',
'DayofWeek',
'OriginAirportID',
'OriginCityName',
'OriginStateName',
'DestAirportID',
'DestCityName',
'DestStateName']]
```

```python
origin = df_flight_status2.select('Origin', 'OriginAirportID', 'OriginCityName', 'OriginStateName').distinct()
origin = origin.withColumnRenamed('Origin', 'id')\
.withColumnRenamed('OriginAirportID', 'AirportID')\
.withColumnRenamed('OriginCityName', 'CityName')\
.withColumnRenamed('OriginStateName', 'StateName')

dest = df_flight_status2.select('Dest', 'DestAirportID', 'DestCityName', 'DestStateName').distinct()
dest = dest.withColumnRenamed('Dest', 'id')\
.withColumnRenamed('DestAirportID', 'AirportID')\
.withColumnRenamed('DestCityName', 'CityName')\
.withColumnRenamed('DestStateName', 'StateName')
```

```python
vertices = origin.union(dest).dropDuplicates(['id'])

edges = df_flight_status2[['Origin', 'Dest', 'Cancelled', 'Diverted', 'Distance', 'DepDelayMinutes']]\
.withColumnRenamed('Origin', 'src')\
.withColumnRenamed('Dest', 'dst')
```

```python
vertices.count()
```

```python
edges.count()
```

```python
edges_sample = edges.sample(1.0)
```

```python
airline_routes = GraphFrame(vertices, edges)
```

```python
def plot_undirected_graph(edge_list):
    plt.figure(figsize = (15, 15))
    gplot = nx.Graph()
    for row in edge_list.select("src", "dst").take(1000):
        gplot.add_edge(row["src"], row["dst"])
    nx.draw(gplot, with_labels = True, font_weight = "bold", node_size = 3500)
plot_undirected_graph(airline_routes.edges)
```

```python
def plot_directed_graph(edge_list):
    plt.figure(figsize=(20, 20))
    gplot = nx.DiGraph()
    for row in edge_list.select("src", "dst").take(10000):
        gplot.add_edge(row["src"], row["dst"])
    nx.draw(gplot, with_labels = True, font_weight = "bold", node_size = 1000)
plot_directed_graph(airline_routes.edges)
```

```python
results = airline_routes.pageRank(resetProbability = 0.15, tol = 0.01)
```

```python
results.vertices.orderBy('pagerank', ascending = False).show()
```

```python
results =  airline_routes.pageRank(resetProbability = 0.15, maxIter = 10, sourceId = "ORD")
results.vertices.orderBy('pagerank', ascending = False).show()
```

```python
paths = airline_routes.shortestPaths(landmarks = ("ORD", "MDW", "AUS"))
paths.show()
```

```python
airline_routes.edges\
.filter("src = 'ORD' and Distance > 1000")\
.groupBy("src", "dst")\
.avg("Distance")\
.orderBy("avg(Distance)", ascending = False)\
.show()
```

```python
airline_routes.edges\
.filter("src = 'ORD' and DepDelayMinutes > 0")\
.groupBy("src", "dst")\
.avg("DepDelayMinutes")\
.orderBy("avg(DepDelayMinutes)", ascending = False)\
.show()
```
